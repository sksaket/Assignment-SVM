```python
# importing the libraries
import numpy as np
from sklearn.svm import SVC
import pandas as pd
```

```python
# reading the data from csv
cols = pd.read_csv("/content/SalaryData_Train(1).csv", nrows = 1).columns
x_train = pd.read_csv("/content/SalaryData_Train(1).csv", usecols = cols[:-1]).to_numpy()
y_train = pd.read_csv("/content/SalaryData_Train(1).csv", usecols = cols[-1:]).to_numpy()
cols = pd.read_csv("/content/SalaryData_Test(1).csv", nrows = 1).columns
x_test = pd.read_csv("/content/SalaryData_Test(1).csv", usecols = cols[:-1]).to_numpy()
y_test = pd.read_csv("/content/SalaryData_Test(1).csv", usecols = cols[-1:]).to_numpy()
```

```python
#this funciton is used to preprocess the data
def preprocess(data):
    data = data.T
    X_trainData = list()
    for row in data:
        l = list()
        temp = list()
        doit = isinstance(row[0], str)
        for column in row:
            if(doit):
                if column not in l:
                    l.append(column)
                temp.append(l.index(column))
            else:
                temp.append(column)
        X_trainData.append(temp)
    return np.array(X_trainData).T
```

```python
#preprocess the data using the above function and the ravel() is used to convert the dimensions from (n, 1) to (n,)
X = preprocess(x_train)
Y = (y_train == " <=50K")
Xtest = preprocess(x_test)
Ytest = (y_test == " <=50K")
Y, Ytest = Y.ravel(), Ytest.ravel()
```

```python
#assign the svc classifier to clf
clf = SVC(gamma = 'auto')
#fit the data to classifier clf
clf.fit(X, Y)
#now check how much accuracy it is giving
score = clf.score(Xtest, Ytest)*100
print("The accuracy of the this is: ", score)
```

```python
#function to predict the values
def predict(Xtest):
    p = clf.predict(Xtest)
    ret = list()
    for i in p:
        if(i):
            ret.append(" <=50K")
        else:
            ret.append(' >50K')
    return ret
#predict the values for test
predict(Xtest)
```
